```python
import time
import requests
import numpy as np
import pandas as pd 
from bs4 import BeautifulSoup
import re
from sqlalchemy import create_engine, text
```

```python
Titles = []
Times = []
Venues =[]
detailNames = []
URLs = []
id = []
```

```python
# id 001
id.append("001")
url = 'https://foiks2024.github.io/'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

title_container = soup.find('font', size = "6")
Titles.append(title_container.text)

time_venue_container = soup.find_all('font', size = "3")[1]
time = time_venue_container.text.split("•")[0]
venue = time_venue_container.text.split("•")[1]
Times.append(time)
Venues.append(venue)

detail_name_container = soup.find_all('font', size = "3")[0]
detailNames.append(detail_name_container.text)
URLs.append(url)

```

```python
# id 015
id.append("015")
url = 'https://www.ecai2024.eu/'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

detail_name = soup.find('h1').text.strip()
time_venue = soup.find('p').text.strip()
time_venue = re.split(r'(2024)', time_venue)

time = time_venue[0] + time_venue[1]
venue = time_venue[2]
title = soup.find('h1', class_ = "__className_8fa1a3").text.strip()

Titles.append(title)
Times.append(time)
Venues.append(venue)
detailNames.append(detail_name)
URLs.append(url)
```
